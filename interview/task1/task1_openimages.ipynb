````python
import pandas as pd
cls=pd.read_csv('oidv6-class-descriptions.csv', header=None)
classes2name={i:j for i,j in zip(cls[0], cls[1])}
name2classes={j:i for i,j in zip(cls[0], cls[1])}
cls.tail()
````

````python
import json
hier = json.load(open('bbox_labels_600_hierarchy.json','r'))
siblings=[]
parents=[]
ancestors=[]
for l2 in hier['Subcategory']:
    print(classes2name[l2['LabelName']])
    ancestors.append(classes2name[l2['LabelName']])
    try:
        for j in l2['Subcategory']:
            print('----> ',classes2name[j['LabelName']])
            parents.append(classes2name[j['LabelName']])
            try:
                for k in j['Subcategory']:
                    print('\t----> ',classes2name[k['LabelName']])
                    siblings.append(classes2name[k['LabelName']])
            except:
                pass
    except:
        pass
        
siblings = set(siblings)
parents = set(parents)
ancestors = set(ancestors)
````

So we got the heirarchy of ancestors, parents and siblings in the above code output.

````python
print('Classes count in siblings is {}'.format(len(siblings)))
print('Classes count in parent is {}'.format(len(parents)))
print('Classes count in ancestors is {}'.format(len(ancestors)))
print('Total unique class counts are {}'.format(len(siblings)+len(parents)+len(ancestors)))
````

````python
print('siblings and parents overlaps = {}'.format(len(parents&siblings)))
print('parents and ancestors overlaps = {}'.format(len(parents&ancestors)))
print('ancestors and siblings overlaps = {}'.format(len(ancestors&siblings)))
````

````python
union_set_parents_siblings = set.union(set(parents), set(siblings))
print(union_set_parents_siblings)
````

````python
final_intersection = set.intersection(set(ancestors), (union_set_parents_siblings))
print(final_intersection)
````

As we can see that we have got a null set, which means that there are no parents class and sibling classes which is also a ancestor class.
